### La matrice , les variables et individus :

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# ============================================================
#  DONNÉES  (matrice du TP1 – 20 villes × 6 sports)
# ============================================================
X = np.array([
    [1881.9, 96.8,  14.2, 25.2, 1135.5, 278.3],
    [3369.8, 96.8,  10.8, 51.6, 1331.7, 284.0],
    [4467.4, 138.2,  9.5, 34.2, 2346.1, 312.3],
    [1862.1, 83.2,   8.8, 27.6,  972.6, 203.4],
    [3499.8, 287.0, 11.5, 49.4, 2139.4, 358.0],
    [3903.2, 170.7,  6.3, 42.0, 1935.2, 292.9],
    [2620.7, 129.5,  4.2, 16.8, 1346.0, 131.8],
    [3678.4, 157.0,  6.0, 24.9, 1682.6, 194.2],
    [3840.5, 187.9, 10.2, 39.6, 1859.9, 449.1],
    [2170.2, 140.5, 11.7, 31.1, 1351.1, 256.5],
    [3920.4, 128.0,  7.2, 25.5, 1911.5,  64.1],
    [2599.6,  39.6,  5.5, 19.4, 1050.8, 172.5],
    [2828.5, 211.3,  9.9, 21.8, 1085.0, 209.0],
    [2498.7, 123.2,  7.4, 26.5, 1086.2, 153.5],
    [2685.1,  41.2,  2.3, 10.6,  812.5,  89.8],
    [2739.3, 100.7,  6.6, 22.0, 1270.4, 180.5],
    [1662.1,  81.1, 10.1, 19.1,  872.2, 123.3],
    [2469.9, 142.9, 15.5, 30.9, 1165.5, 335.5],
    [2350.7,  38.7,  2.4, 13.5, 1253.1, 170.0],
    [3177.7, 292.1,  8.0, 34.8, 1400.0, 358.9],
])

villes = [f"V{i}" for i in range(1, 21)]
sports = ["H.Ball", "B.Ball", "Tennis", "Gym", "Natation", "F.Ball"]

m, n = X.shape      # m=20 individus, n=6 variables
#OUTPUT_DIR = os.path.dirname(os.path.abspath(__file__))
```

### PRÉ-TRAITEMENT  

Centrage-réduction : Z[i,j] = (X[i,j] - mean_j) / std_j

```python
mean = np.mean(X, axis=0)
std  = np.std(X, axis=0, ddof=0)
Z    = (X - mean) / std
```

Matrice de corrélation : R = (1/m) * Z^T * Z

```python
R = (1 / m) * Z.T @ Z
```

valeurs propres et vecteurs propres

```python
eigenvalues, eigenvectors = np.linalg.eigh(R)
idx         = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
```

Inertie (%) de chaque axe

```python
inertia     = eigenvalues / eigenvalues.sum() * 100
cum_inertia = np.cumsum(inertia)
```

Nombre d'axes retenus (λ ≥ 1)

```python
q = int(np.sum(eigenvalues >= 1.0))   # q = 2
```

Projections de TOUS les individus sur TOUS les axes

```python
F_all = Z @ eigenvectors   # (20 × 6)
F     = F_all[:, :q]       # (20 × 2) — uniquement F1 et F2

axes_labels = [f'F{i+1}' for i in range(n)]
```

Fonctions utilitaires d'affichage

```python
def sep(title):
    print("\n" + "=" * 65)
    print(title)
    print("=" * 65)

def print_table_header(cols, col_w=12):
    print("  ".join(f"{c:>{col_w}}" for c in cols))
    print("-" * (col_w * len(cols) + 2 * (len(cols) - 1)))

def print_table_row(vals, col_w=12, decimals=4):
    row = ""
    for v in vals:
        if isinstance(v, str):
            row += f"{v:>{col_w}}  "
        else:
            row += f"{v:>{col_w}.{decimals}f}  "
    print(row)
```

### A.Q1 – Carré de la distance de chaque individu au centre dans le sous-espace factoriel

Methode :
en va calculer le distance entre l'origine de sous - espace factorielle retenu et chaque individu ,

donc en vas utiliser la methode de Pythagore :

la coordoner d'individue sur F1 au caree + la coordoner d'individue sur F2 au caree = la distance au caree

la distance = sqrt(la coordoner d'individue sur F1 au caree + la coordoner d'individue sur F2 au caree)

```python

sep("A.Q1 – CARRÉ DE LA DISTANCE AU CENTRE (sous-espace factoriel)")

print("\nFormule : d²(i) = Σ_{k=1}^{q} F[i,k]²  =  F1(i)²  +  F2(i)²")
print("          (Pythagore dans le plan F1×F2, origine = centre du nuage)")
print()

# d²(i) = somme des carrés des coordonnées sur les q axes retenus
d2 = np.sum(F**2, axis=1)   # tableau de 20 valeurs, une par ville

col_w = 16
print_table_header(["Ville", "F1(i)²", "F2(i)²", "d²(i) = F1²+F2²"], col_w)
for i in range(m):
    print_table_row([villes[i], F[i,0]**2, F[i,1]**2, d2[i]], col_w)
```

### A.Q2 – Distances des individus + somme totale

```python
sep("A.Q2 – DISTANCES DES INDIVIDUS + SOMME")

print("\nFormule : d(i) = √( d²(i) )  =  √( F1(i)² + F2(i)² )")
print()

d = np.sqrt(d2)

col_w = 16
print_table_header(["Ville", "d²(i)", "d(i)"], col_w)
for i in range(m):
    print_table_row([villes[i], d2[i], d[i]], col_w)

somme_d2 = d2.sum()
somme_d = d.sum()
valeur_theorique = m * eigenvalues[:q].sum()
print(f"\n  Somme des d(i)  =  {somme_d:.4f}")
```

### A.Q4 – Qualité de représentation des individus (cos²)

```python

sep("A.Q4 – QUALITÉ DE REPRÉSENTATION DES INDIVIDUS (cos²)")

print("\nFormule : cos²(i, k) = F_all[i,k]²  /  d²_total(i)")
print("          avec d²_total(i) = Σ_{k=1}^{n} F_all[i,k]²  (distance dans l'espace complet)")
print("          cos²(i,F1) + cos²(i,F2) + ... + cos²(i,F6) = 1  (propriété de vérification)")
print()

# Distance au carré dans l'espace COMPLET (tous les 6 axes)
# nécessaire pour normaliser le cos²
d2_total = np.sum(F_all**2, axis=1)   # (20,)

# cos²[i,k] = F_all[i,k]² / d²_total[i]
cos2 = F_all**2 / d2_total[:, np.newaxis]   # (20 × 6)

col_w = 12
print_table_header(["Ville", "cos²(F1)", "cos²(F2)", "cos²(F3)",
                    "cos²(F1+F2)", "Qualité plan"], col_w)
for i in range(m):
    c1   = cos2[i, 0]
    c2   = cos2[i, 1]
    c3   = cos2[i, 2]
    plan = c1 + c2
    if plan >= 0.80:
        qual = "Très bonne"
    elif plan >= 0.60:
        qual = "Bonne"
    elif plan >= 0.40:
        qual = "Moyenne"
    else:
        qual = "Mauvaise"
    print_table_row([villes[i], c1, c2, c3, plan, qual], col_w)

print(f"""
Interprétation :
  • cos²(i,k) proche de 1 → l'individu est quasi entièrement porté
    par cet axe : position significative dans cette direction.
  • cos²(i, F1+F2) ≥ 0.80 → très bonne représentation dans le plan.
  • cos²(i, F1+F2) < 0.50 → individu mal représenté dans le plan :
    sa position dans le graphe n'est pas fiable, ne pas l'interpréter.
  • La colonne cos²(F3) indique si l'axe F3 (non retenu) apporte
    de l'information supplémentaire pour cet individu.
""")
```

### A.Q5 – Contribution des individus à l'inertie de chaque axe

#### MÉTHODE :
  La contribution de l'individu i à l'axe k mesure sa part dans la construction (maximisation de la variance) de cet axe:

     CTR(i, k) = F_all[i,k]²  /  (m × λk)

  où m = nombre d'individus, λk = valeur propre de l'axe k.


```python

sep("A.Q5 – CONTRIBUTION DES INDIVIDUS À L'INERTIE DES AXES (CTR)")

print("\nFormule : CTR(i, k) = F_all[i,k]²  /  (m × λk)")
print(f"          Seuil de référence = 1/m = 1/{m} = {1/m:.4f} = {100/m:.2f}%")
print(f"          (contribution 'normale' si tous les individus sont équivalents)")
print()

# CTR[i,k] = F_all[i,k]² / (m * λk)
CTR = F_all**2 / (m * eigenvalues[np.newaxis, :])   # (20 × 6)

col_w = 12
print_table_header(["Ville", "CTR(F1)%", "CTR(F2)%", "CTR(F3)%"], col_w)
for i in range(m):
    print_table_row([villes[i],
                     CTR[i, 0] * 100,
                     CTR[i, 1] * 100,
                     CTR[i, 2] * 100], col_w)

print(f"\n  Vérification – Somme CTR(F1) = {CTR[:,0].sum()*100:.2f}%"
      f"  | CTR(F2) = {CTR[:,1].sum()*100:.2f}%"
      f"  | CTR(F3) = {CTR[:,2].sum()*100:.2f}%")

print(f"""
Interprétation :
  • Seuil de référence = {100/m:.2f}% par axe.
  • CTR >> {100/m:.1f}% → individu très influent sur la construction de l'axe.
    Il "attire" l'axe vers lui : sa position dans le graphe est structurante.
  • CTR << {100/m:.1f}% → individu passif, peu d'influence sur l'axe.
  • Un individu à la fois loin du centre (d² grand) ET à forte CTR
    est un individu "atypique actif" : il influence fortement les axes.
  • La somme de chaque colonne = 100% (propriété de vérification).
""")
```

### A.Q6 – Individus mal représentés + signe sur chaque axe

##### MÉTHODE :
  Un individu est mal représenté sur l'axe k si cos²(i,k) est faible.

  Seuil retenu : cos²(i,k) < 0.10  (contribution < 10% de la distance).


```python
sep("A.Q6 – INDIVIDUS MAL REPRÉSENTÉS SUR CHAQUE AXE")

print("\nSeuil : cos²(i,k) < 0.10  →  individu mal représenté sur l'axe k")
print("Signe : indiqué par le signe de la coordonnée F_all[i,k]")
print()

seuil = 0.10
print(f"{'Ville':<10}  {'Axe':<5}  {'cos²':>8}  {'Coordonnée':>12}  {'Signe':>12}  Verdict")
print("-" * 68)

found = False
for i in range(m):
    for k in range(q):     # on examine uniquement les 2 axes retenus
        c2 = cos2[i, k]
        if c2 < seuil:
            coord = F_all[i, k]
            signe = "positif (+)" if coord > 0 else "négatif (−)"
            found = True
            print(f"{villes[i]:<10}  F{k+1:<4}  {c2:>8.4f}  {coord:>12.4f}  {signe:>12}  ⚠ Mal représenté")

if not found:
    print("  Aucun individu n'est mal représenté (cos² ≥ 0.10) sur les axes retenus.")

print(f"""
Remarque :
  • Un individu mal représenté sur F1 ET F2 ne doit PAS être interprété
    dans le graphe : sa position dans le plan est due au hasard de la projection.
  • Un individu mal représenté sur F1 seulement peut encore être interprété
    selon son axe F2 (et inversement).
  • Le signe indique de quel côté l'individu semble se trouver,
    mais cette information n'est pas fiable si cos² est trop faible.
""")
```
